```python
import json
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('font',family='Times New Roman')
plt.rcParams['font.weight'] = 'bold'

evol_name = 'MRD$^3$ w/o Evol'
#f = open('./score_{}_math10k_revised.json'.format(evol_name))
f = open('./score_math10k_v2.json')
cot_data = json.load(f) 
f.close()

difficulty_list = []
steps_list = []

for i in tqdm(range(len(cot_data))):
    difficulty = int(cot_data[i]['difficulty'])
    steps = int(cot_data[i]['steps'])
    difficulty_list.append(difficulty)
    steps_list.append(steps)

level = [str(x) for x in range(1,11)]
difficulty_count = [0]*10
for difficulty in difficulty_list:
    difficulty_count[difficulty-1] += 1

plt.figure(figsize=(5,4))
plt.bar(level,difficulty_count,color='tab:blue', alpha=0.8)
plt.xticks(fontsize=16)
plt.yticks([0,1000,2000,3000,4000,5000], fontsize=16)
plt.ylim(0,5000)
plt.xlabel("Difficulty Level of {}".format(evol_name), fontsize=16)
plt.grid(linestyle='dotted')
plt.savefig("./plot_figs/difficulty_no_evol.pdf",dpi=200,bbox_inches = 'tight')
plt.show()

level = [str(x) for x in range(1,10)]
step_count = [0]*9
for steps in steps_list:
    step_count[steps-1] += 1

plt.figure(figsize=(5,4))
plt.bar(level,step_count,color='tab:orange', alpha=0.8)
plt.xticks(fontsize=16)
plt.yticks([0,1000,2000,3000,4000], fontsize=16)
plt.ylim(0,4000)
plt.xlabel("Reasoning Steps of {}".format(evol_name), fontsize=16)
plt.grid(linestyle='dotted')
plt.savefig("./plot_figs/step_no_evol.pdf",dpi=200,bbox_inches = 'tight')
plt.show()
```
